# Applying Operators 

Sometimes it is difficult to deal with data on some hypothesis because of some random data in our dataset.The hypothesis will not work on this type of data. It will generate very bad results.<br>
Like we have some data of room temperature(degrees) of a given period as given below<br>
<b><code>data = {'temperatures':[25.58, 25.67, 26.50, 27.50, 27.55, 28.26,28.26 28.27, <u>8445.45</u>, 27.50, 26.50, 25.45, 24.21, 22.48]}</code></b><br>
We can clearly see that 8445.45 can't be the data point of our set.This is because of some random error of instrumentation error.<br>
So now our task is to filter these type of data from our dataset so that we can easily apply some algorithm to predict or analysis data.

```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")
```

```python
data = data = {'temperatures':[25.58, 25.67, 26.50, 27.50, 27.55, 28.26,28.26, 28.27, 8445.45, 27.50, 26.50, 25.45, 24.21, 22.48]}
df = pd.DataFrame(data)
# print(df['temperatures'])
```

```python
df.plot()
plt.show()
##here we can clealy see the change 
```

```python
# getting rolling STD in window of 2
df['STD'] = pd.Series(df['temperatures']).rolling(window=2).std()
print(df)
```

```python
df.plot()
plt.show()
## there is an abrupt change in standard deviation from 6 to 9
```

```python
STD_all = df['temperatures'].std()
print(STD_all)
```

```python
##now we will create a new dataframe in which all the values have STD less than STD_all
df_main = df[( df['STD'] < STD_all )]
print(df_main)
##finally we have got rid from random values
```

```python
df_main['temperatures'].plot()
plt.show()
##now this graph makes some sense, We can easily Quardic Regression on this data frame
```
